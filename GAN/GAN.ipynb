```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import torchvision.utils as vutils
import matplotlib.pyplot as plt
from pytorch_fid import fid_score

# ---------------------------
# 1. Define the Models
# ---------------------------
class Generator(nn.Module):
    def __init__(self, latent_dim=100):
        super(Generator, self).__init__()
        self.model = nn.Sequential(
            # Input: latent vector Z (latent_dim, 1, 1)
            nn.ConvTranspose2d(latent_dim, 256, kernel_size=7, stride=1, padding=0, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(True),
            # Shape: (256, 7, 7)
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            # Shape: (128, 14, 14)
            nn.ConvTranspose2d(128, 1, kernel_size=4, stride=2, padding=1, bias=False),
            nn.Tanh()  # Final shape: (1, 28, 28)
        )

    def forward(self, z):
        return self.model(z)

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            # Input: (1, 28, 28)
            nn.Conv2d(1, 64, kernel_size=4, stride=2, padding=1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),
            # Shape: (64, 14, 14)
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
            # Shape: (128, 7, 7)
            nn.Flatten(),
            nn.Linear(128 * 7 * 7, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.model(x)

def weights_init(m):
    classname = m.__class__.__name__
    if 'Conv' in classname or 'Linear' in classname:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
        if m.bias is not None:
            nn.init.constant_(m.bias.data, 0)

# ---------------------------
# 2. Hyperparameters and Data Loading
# ---------------------------
latent_dim = 100
batch_size = 128
lr = 0.0002
epochs = 50  # Increased to 50 epochs
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Directories for saving images
os.makedirs("samples", exist_ok=True)
os.makedirs("real_images", exist_ok=True)

# Data transforms and dataset
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
dataset = datasets.MNIST(root="./data", train=True, transform=transform, download=True)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# ---------------------------
# 3. Prepare Real Images for FID
# ---------------------------
if not os.listdir("real_images"):
    print("Saving real images for FID computation...")
    real_loader = DataLoader(dataset, batch_size=100, shuffle=True)
    count = 0
    for imgs, _ in real_loader:
        for j in range(imgs.size(0)):
            vutils.save_image(imgs[j], f"real_images/real_{count:05d}.png", normalize=True)
            count += 1
            if count >= 1000:
                break
        if count >= 1000:
            break

# ---------------------------
# 4. Training Function with Epoch Loss Display
# ---------------------------
def train_gan(discriminator_update_steps):
    # Initialize models
    generator = Generator(latent_dim).to(device)
    discriminator = Discriminator().to(device)
    generator.apply(weights_init)
    discriminator.apply(weights_init)

    adversarial_loss = nn.BCELoss()
    optimizer_G = optim.Adam(generator.parameters(), lr=lr, betas=(0.5, 0.999))
    optimizer_D = optim.Adam(discriminator.parameters(), lr=lr, betas=(0.5, 0.999))

    epoch_losses_D = []
    epoch_losses_G = []

    for epoch in range(epochs):
        running_loss_D = 0.0
        running_loss_G = 0.0
        num_batches = 0
        for i, (real_imgs, _) in enumerate(dataloader):
            real_imgs = real_imgs.to(device)
            current_batch_size = real_imgs.size(0)
            num_batches += 1

            # Train Discriminator (multiple steps if needed)
            for _ in range(discriminator_update_steps):
                discriminator.zero_grad()
                real_labels = torch.full((current_batch_size, 1), 1.0, device=device)
                output_real = discriminator(real_imgs)
                loss_D_real = adversarial_loss(output_real, real_labels)

                noise = torch.randn(current_batch_size, latent_dim, 1, 1, device=device)
                fake_imgs = generator(noise)
                fake_labels = torch.full((current_batch_size, 1), 0.0, device=device)
                output_fake = discriminator(fake_imgs.detach())
                loss_D_fake = adversarial_loss(output_fake, fake_labels)

                loss_D = loss_D_real + loss_D_fake
                loss_D.backward()
                optimizer_D.step()

            # Train Generator
            generator.zero_grad()
            noise = torch.randn(current_batch_size, latent_dim, 1, 1, device=device)
            fake_imgs = generator(noise)
            target_labels = torch.full((current_batch_size, 1), 1.0, device=device)
            output = discriminator(fake_imgs)
            loss_G = adversarial_loss(output, target_labels)
            loss_G.backward()
            optimizer_G.step()

            running_loss_D += loss_D.item()
            running_loss_G += loss_G.item()

        # Compute average loss for the epoch
        avg_loss_D = running_loss_D / num_batches
        avg_loss_G = running_loss_G / num_batches
        epoch_losses_D.append(avg_loss_D)
        epoch_losses_G.append(avg_loss_G)
        print(f"Epoch [{epoch+1}/{epochs}] | Loss_D: {avg_loss_D:.4f} | Loss_G: {avg_loss_G:.4f}")

    return generator, epoch_losses_D, epoch_losses_G

# ---------------------------
# 5. Generate Fake Images and Compute FID
# ---------------------------
def generate_fake_images(generator, num_images=1000, folder="gen_images"):
    os.makedirs(folder, exist_ok=True)
    generator.eval()
    count = 0
    with torch.no_grad():
        while count < num_images:
            noise = torch.randn(batch_size, latent_dim, 1, 1, device=device)
            fake_imgs = generator(noise).detach().cpu()
            for j in range(fake_imgs.size(0)):
                if count >= num_images:
                    break
                vutils.save_image(fake_imgs[j], f"{folder}/fake_{count:05d}.png", normalize=True)
                count += 1

def display_generated_grid(generator, epoch_label):
    generator.eval()
    with torch.no_grad():
        fixed_noise = torch.randn(16, latent_dim, 1, 1, device=device)
        gen_samples = generator(fixed_noise).detach().cpu()
        grid = vutils.make_grid(gen_samples, padding=2, normalize=True)
    plt.figure(figsize=(4, 4))
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis("off")
    plt.title(f"Generated Samples ({epoch_label})")
    plt.show()

# ---------------------------
# 6. Loop Over Discriminator Update Steps, Display Loss, Grid and Compute FID
# ---------------------------
update_steps_list = [1, 5,10,15]
results = {}

for d_steps in update_steps_list:
    print(f"\n=== Training with discriminator_update_steps = {d_steps} ===")
    generator, losses_D, losses_G = train_gan(d_steps)

    # Display loss curves
    plt.figure()
    plt.plot(range(1, epochs+1), losses_D, label="Discriminator Loss")
    plt.plot(range(1, epochs+1), losses_G, label="Generator Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"Loss Curves (Update Steps = {d_steps})")
    plt.legend()
    plt.show()

    # Display a grid of generated images.
    display_generated_grid(generator, f"Update Steps {d_steps}")

    # Generate fake images for FID computation.
    fake_folder = f"gen_images_ds_{d_steps}"
    if os.path.exists(fake_folder):
        # Clean the folder if it exists.
        for file in os.listdir(fake_folder):
            os.remove(os.path.join(fake_folder, file))
    else:
        os.makedirs(fake_folder)
    generate_fake_images(generator, num_images=500, folder=fake_folder)

    # Compute FID between real images and generated images.
    fid_value = fid_score.calculate_fid_given_paths([os.path.abspath("real_images"), os.path.abspath(fake_folder)],
                                                      batch_size, device, dims=2048)
    print(f"Discriminator Update Steps: {d_steps} --> FID Score: {fid_value:.4f}")
    results[d_steps] = {"FID": fid_value, "loss_D": losses_D, "loss_G": losses_G}

# ---------------------------
# 7. Summary of Results
# ---------------------------
print("\n=== Summary of Results ===")
for d_steps, metrics in results.items():
    print(f"Update Steps: {d_steps} | Final FID: {metrics['FID']:.4f}")
```

# Summary of Discriminator Update Experiments

The following results were obtained when experimenting with different numbers of discriminator update steps per generator update:

- **Update Steps: 1** | Final FID: 27.6996
- **Update Steps: 5** | Final FID: 25.3971
- **Update Steps: 10** | Final FID: 26.3111
- **Update Steps: 15** | Final FID: 51.1630

## Analysis

- **1 Update Step:**  
  With a single update of the discriminator per generator update, the final FID score is **27.6996**. This setting provides a baseline where the discriminator is not overtrained relative to the generator.

- **5 Update Steps:**  
  Increasing the discriminator updates to 5 per generator update improves the FID to **25.3971**. This suggests that a stronger discriminator—trained more often—can provide better gradients to the generator, leading to higher quality outputs (as indicated by a lower FID).

- **10 Update Steps:**  
  At 10 updates, the FID slightly increases to **26.3111**. Although still comparable to the 5-update setting, this change indicates that further increasing the discriminator's strength may not yield additional benefits and might even start to create an imbalance.

- **15 Update Steps:**  
  With 15 discriminator updates per generator update, the FID worsens dramatically to **51.1630**. This is a strong indication that overtraining the discriminator can be detrimental. An excessively strong discriminator might dominate the adversarial game, leading to vanishing gradients for the generator and, in turn, causing poor generation quality.

## Conclusion

There is a clear trade-off in setting the number of discriminator updates:
- A moderate number (around 5) helps the generator by providing robust feedback, thereby improving sample quality.
- Too few updates leave the discriminator under-trained, while too many updates (e.g., 15) can destabilize training and harm the generator's performance.

Thus, the optimal balance appears to be achieved with around **5 updates per generator step** for this experiment, as it yielded the lowest FID score.


# Comparison of GAN, VAE, and DDPM Samples

After training each model for 50 epochs on the same dataset, we observed the following qualitative and quantitative differences:

## Qualitative Comparison

- **Variational Autoencoder (VAE):**
  - **Smooth and Consistent:**  
    The VAE produced smooth, well-structured samples. Its probabilistic framework and latent space regularization promote smooth interpolations and consistent reconstructions.
  - **High Diversity:**  
    The generated digits were clear and diverse, consistently matching the style of real images.

- **Denoising Diffusion Probabilistic Model (DDPM):**
  - **High Fidelity and Detail:**  
    DDPM generated images with excellent sharpness and fine details, thanks to its iterative denoising process.
  - **Slightly Reduced Diversity:**  
    Although the details are impressive, the overall diversity of the samples was slightly lower compared to the VAE.

- **Generative Adversarial Network (GAN):**
  - **Artifacts and Instabilities:**  
    GAN samples sometimes exhibited visual artifacts and were prone to issues like mode collapse. This is common due to the sensitivity of GAN training dynamics.
  - **Lower Overall Consistency:**  
    While some GAN samples were appealing, the overall consistency and quality were lower than those from the VAE and DDPM.

## Quantitative Comparison

Using the Fréchet Inception Distance (FID) metric:
- **VAE:**  
  Achieved the lowest FID score, indicating that its generated samples are closest in distribution to the real images.
- **DDPM:**  
  Performed slightly worse than the VAE but still significantly better than the GAN.
- **GAN:**  
  Registered the highest FID score, reflecting the training challenges and inconsistencies (e.g., mode collapse) observed during qualitative evaluation.

## Explanation

- **Why the VAE Performs Well:**  
  The VAE's design incorporates a strong regularization of its latent space, encouraging smooth and continuous representations. This leads to high-quality, diverse reconstructions that closely follow the true data distribution. Its probabilistic nature helps in capturing a broad range of features, which is reflected in both low FID scores and visually appealing outputs.

- **DDPM's Strengths and Trade-offs:**  
  The DDPM, through its iterative denoising approach, is capable of generating images with remarkable detail and sharpness. However, the iterative process may slightly compromise the diversity of the samples compared to the VAE, which is why its FID score is slightly higher even though the image details are excellent.

- **Challenges with GANs:**  
  GANs are known for their training instability. Issues like mode collapse can cause the generator to produce a limited variety of outputs, which degrades both the qualitative appearance and quantitative FID metrics. The GAN's sensitivity to hyperparameters and the adversarial training setup often results in a higher FID score compared to the VAE and DDPM.

## Conclusion

After 50 epochs of training:
- **VAE** outperforms the other models both qualitatively and quantitatively.
- **DDPM** follows, delivering detailed images but with somewhat reduced diversity.
- **GAN** trails behind due to inherent training instabilities and mode collapse issues.

This order — **VAE > DDPM > GAN** — is reflected in both the FID scores and the visual quality of the generated samples.


```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import torchvision.utils as vutils
import matplotlib.pyplot as plt
from pytorch_fid import fid_score

# ---------------------------
# 1. Define the Models
# ---------------------------
class Generator(nn.Module):
    def __init__(self, latent_dim=100):
        super(Generator, self).__init__()
        self.model = nn.Sequential(
            # Input: latent vector Z (latent_dim, 1, 1)
            nn.ConvTranspose2d(latent_dim, 256, kernel_size=7, stride=1, padding=0, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(True),
            # Shape: (256, 7, 7)
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            # Shape: (128, 14, 14)
            nn.ConvTranspose2d(128, 1, kernel_size=4, stride=2, padding=1, bias=False),
            nn.Tanh()  # Final shape: (1, 28, 28)
        )

    def forward(self, z):
        return self.model(z)

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            # Input: (1, 28, 28)
            nn.Conv2d(1, 64, kernel_size=4, stride=2, padding=1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),
            # Shape: (64, 14, 14)
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
            # Shape: (128, 7, 7)
            nn.Flatten(),
            nn.Linear(128 * 7 * 7, 1),
            nn.Sigmoid()  # Output probability
        )

    def forward(self, x):
        return self.model(x)

def weights_init(m):
    classname = m.__class__.__name__
    if 'Conv' in classname or 'Linear' in classname:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
        if m.bias is not None:
            nn.init.constant_(m.bias.data, 0)

# ---------------------------
# 2. Hyperparameters and Data Loading
# ---------------------------
latent_dim = 100
batch_size = 128
lr = 0.0002
epochs = 50  # Increase epochs for thorough training
# Overtraining discriminator to encourage mode collapse
discriminator_update_steps = 5

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# MNIST transforms and dataset
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
dataset = datasets.MNIST(root="./data", train=True, transform=transform, download=True)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# ---------------------------
# 3. Prepare Real Images for FID
# ---------------------------
real_images_dir = "real_images"
os.makedirs(real_images_dir, exist_ok=True)
if not os.listdir(real_images_dir):
    print("Saving real images for FID computation...")
    real_loader = DataLoader(dataset, batch_size=100, shuffle=True)
    count = 0
    for imgs, _ in real_loader:
        for j in range(imgs.size(0)):
            vutils.save_image(imgs[j], f"{real_images_dir}/real_{count:05d}.png", normalize=True)
            count += 1
            if count >= 1000:
                break
        if count >= 1000:
            break

# ---------------------------
# 4. Initialize Models, Loss, and Optimizers
# ---------------------------
generator = Generator(latent_dim).to(device)
discriminator = Discriminator().to(device)
generator.apply(weights_init)
discriminator.apply(weights_init)

# Standard BCE loss for the discriminator
adversarial_loss = nn.BCELoss()
# L1 loss for forcing mode collapse in the generator
l1_loss = nn.L1Loss()

optimizer_G = optim.Adam(generator.parameters(), lr=lr, betas=(0.5, 0.999))
optimizer_D = optim.Adam(discriminator.parameters(), lr=lr, betas=(0.5, 0.999))

# Fixed target image for mode collapse – an all-zero image
fixed_target = torch.zeros((1, 1, 28, 28), device=device)

# ---------------------------
# 5. Training Loop (Intentional Mode Collapse)
# ---------------------------
loss_history_D = []
loss_history_G = []

print("Starting GAN Training with Intentional Mode Collapse...")

for epoch in range(epochs):
    epoch_loss_D = 0.0
    epoch_loss_G = 0.0
    num_batches = 0

    for i, (real_imgs, _) in enumerate(dataloader):
        real_imgs = real_imgs.to(device)
        current_batch_size = real_imgs.size(0)
        num_batches += 1

        # --- Train Discriminator: Overtrain it ---
        for _ in range(discriminator_update_steps):
            discriminator.zero_grad()
            # Real images
            real_labels = torch.full((current_batch_size, 1), 1.0, device=device)
            output_real = discriminator(real_imgs)
            loss_D_real = adversarial_loss(output_real, real_labels)

            # Fake images
            noise = torch.randn(current_batch_size, latent_dim, 1, 1, device=device)
            fake_imgs = generator(noise)
            fake_labels = torch.full((current_batch_size, 1), 0.0, device=device)
            output_fake = discriminator(fake_imgs.detach())
            loss_D_fake = adversarial_loss(output_fake, fake_labels)

            loss_D = loss_D_real + loss_D_fake
            loss_D.backward()
            optimizer_D.step()

        # --- Train Generator: Using L1 loss to force mode collapse ---
        generator.zero_grad()
        noise = torch.randn(current_batch_size, latent_dim, 1, 1, device=device)
        fake_imgs = generator(noise)
        loss_G = l1_loss(fake_imgs, fixed_target.expand_as(fake_imgs))
        loss_G.backward()
        optimizer_G.step()

        epoch_loss_D += loss_D.item()
        epoch_loss_G += loss_G.item()

    avg_loss_D = epoch_loss_D / num_batches
    avg_loss_G = epoch_loss_G / num_batches
    loss_history_D.append(avg_loss_D)
    loss_history_G.append(avg_loss_G)
    print(f"Epoch [{epoch+1}/{epochs}] - Avg Loss D: {avg_loss_D:.4f}, Avg Loss G: {avg_loss_G:.4f}")

# ---------------------------
# 6. Display Generated Images and Loss Plots
# ---------------------------
def display_generated_grid(generator, title="Generated Samples"):
    generator.eval()
    with torch.no_grad():
        fixed_noise = torch.randn(16, latent_dim, 1, 1, device=device)
        gen_samples = generator(fixed_noise).detach().cpu()
        grid = vutils.make_grid(gen_samples, padding=2, normalize=True)
    plt.figure(figsize=(4, 4))
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis("off")
    plt.title(title)
    plt.show()

display_generated_grid(generator, title="Final Generated Samples (Mode Collapse)")

plt.figure(figsize=(10,5))
plt.plot(range(1, epochs+1), loss_history_D, label="Discriminator Loss")
plt.plot(range(1, epochs+1), loss_history_G, label="Generator Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Curves during Mode Collapse Training")
plt.legend()
plt.show()

# ---------------------------
# 7. Generate Fake Images and Compute FID
# ---------------------------
def generate_fake_images(generator, num_images=1000, folder="gen_images_mode_collapse"):
    os.makedirs(folder, exist_ok=True)
    generator.eval()
    count = 0
    with torch.no_grad():
        while count < num_images:
            noise = torch.randn(batch_size, latent_dim, 1, 1, device=device)
            fake_imgs = generator(noise).detach().cpu()
            for j in range(fake_imgs.size(0)):
                if count >= num_images:
                    break
                vutils.save_image(fake_imgs[j], f"{folder}/fake_{count:05d}.png", normalize=True)
                count += 1

fake_images_dir = "gen_images_mode_collapse"
# Clear the directory if it exists
if os.path.exists(fake_images_dir):
    for file in os.listdir(fake_images_dir):
        os.remove(os.path.join(fake_images_dir, file))
else:
    os.makedirs(fake_images_dir)

print("Generating fake images for FID computation...")
generate_fake_images(generator, num_images=500, folder=fake_images_dir)

print("Computing FID score for mode collapse outputs...")
fid_value = fid_score.calculate_fid_given_paths([os.path.abspath(real_images_dir), os.path.abspath(fake_images_dir)],
                                                  batch_size, device, dims=2048)
print(f"FID Score for Mode Collapse GAN: {fid_value:.4f}")
```


# Explanation of Mode Collapse in Our GAN

**Mode collapse** occurs when the generator in a GAN learns to produce only a few outputs or even a single output regardless of the variation in its input noise. This results in a lack of diversity in the generated samples and is typically reflected in very high quantitative metrics like the FID score.

## How We Intentionally Induced Mode Collapse

In our experiment, we intentionally introduced mode collapse by combining two strategies:

1. **Overtraining the Discriminator:**  
   - We updated the discriminator multiple times per generator update (e.g., 20 updates per generator update).  
   - This causes the discriminator to become very strong compared to the generator. As a result, the generator is forced to converge quickly to a narrow set of outputs that are “good enough” to fool the discriminator, rather than exploring the full data distribution.

2. **Modifying the Generator Loss:**  
   - Instead of using the standard adversarial loss, we replaced it with an L1 loss that forces the generator’s output to be close to a fixed target (for example, an all-zero image).  
   - This modification eliminates the incentive for the generator to capture diverse features from the data and makes it produce nearly identical outputs regardless of the noise input.

## Observations from Our Experiment

- **High FID Score:**  
  The mode collapse GAN achieved a very high FID score of **355.1271**. This is a strong indicator of both poor image quality and a significant lack of diversity in the generated samples.
  
- **Loss of Diversity:**  
  The generator’s outputs were nearly identical or limited to only a few variations, confirming that it had collapsed to a single mode (or very few modes) instead of covering the full range of possibilities.

## Conclusion

By overtraining the discriminator and replacing the generator’s adversarial loss with a fixed L1 loss, we intentionally induced mode collapse in our GAN. This approach forced the generator to converge on producing only a narrow subset of outputs, leading to a dramatic increase in the FID score and a clear loss of output diversity.

```python
import os
import torch
import clip
from torchvision import transforms, datasets
from torch.utils.data import DataLoader, Subset
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# ---------------------------
# 1. Load the CLIP model ("ViT-L/14") and create text embeddings
# ---------------------------
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess_clip = clip.load("ViT-L/14", device=device)

# Define class names and corresponding text prompts
class_names = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
text_prompts = [f"the handwritten digit {name}" for name in class_names]
with torch.no_grad():
    text_tokens = clip.tokenize(text_prompts).to(device)
    text_embeddings = model.encode_text(text_tokens)
    text_embeddings /= text_embeddings.norm(dim=-1, keepdim=True)

# ---------------------------
# 2. Load and preprocess the MNIST test dataset
# ---------------------------
# CLIP expects images sized 224x224 with 3 channels.
transform_mnist = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.Grayscale(num_output_channels=3),  # Convert grayscale to 3 channels
    transforms.ToTensor(),
    transforms.Normalize((0.48145466, 0.4578275, 0.40821073),
                         (0.26862954, 0.26130258, 0.27577711))  # CLIP normalization
])
mnist_test = datasets.MNIST(root="./data", train=False, transform=transform_mnist, download=True)
test_loader = DataLoader(mnist_test, batch_size=128, shuffle=False)

# ---------------------------
# 3. Zero-shot Classification Accuracy on MNIST test set
# ---------------------------
all_preds = []
all_labels = []

model.eval()
with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        image_embeddings = model.encode_image(images)
        image_embeddings /= image_embeddings.norm(dim=-1, keepdim=True)
        similarity = image_embeddings @ text_embeddings.T  # (batch_size, num_classes)
        preds = similarity.argmax(dim=-1).cpu().numpy()
        all_preds.extend(preds.tolist())
        all_labels.extend(labels.numpy().tolist())

accuracy = np.mean(np.array(all_preds) == np.array(all_labels))
print(f"Zero-shot CLIP Accuracy on MNIST test set using ViT-L/14: {accuracy * 100:.2f}%")

# ---------------------------
# 4. Filter and Visualize "Best Matched" Images using t-SNE
# ---------------------------
# Select a random subset (e.g., 1000 images) from the test set.
subset_size = 1000
indices = np.random.choice(len(mnist_test), subset_size, replace=False)
subset = Subset(mnist_test, indices)
subset_loader = DataLoader(subset, batch_size=128, shuffle=False)

# Define a threshold for "best matched" images (high confidence)
similarity_threshold = 0.25  # Adjust this value as needed

best_embeddings = []
best_preds = []
best_labels = []
all_max_scores = []

with torch.no_grad():
    for images, labels in subset_loader:
        images = images.to(device)
        emb = model.encode_image(images)
        emb /= emb.norm(dim=-1, keepdim=True)
        sim = emb @ text_embeddings.T  # Compute cosine similarities
        max_scores, preds = sim.max(dim=-1)

        # Create a filter mask for high confidence predictions
        filter_mask = max_scores > similarity_threshold
        if filter_mask.sum().item() > 0:
            best_embeddings.append(emb[filter_mask].cpu().numpy())
            best_preds.extend(preds[filter_mask].cpu().numpy().tolist())
            # Move filter_mask to CPU before indexing labels
            best_labels.extend(labels[filter_mask.cpu()].numpy().tolist())
            all_max_scores.extend(max_scores[filter_mask].cpu().numpy().tolist())

if best_embeddings:
    best_embeddings = np.concatenate(best_embeddings, axis=0)
else:
    print("No images passed the confidence threshold.")
    best_embeddings = np.array([])

print(f"Number of best matched images: {best_embeddings.shape[0]}")

# ---------------------------
# 5. t-SNE Visualization of the Best Matched Images
# ---------------------------
if best_embeddings.shape[0] > 0:
    tsne = TSNE(n_components=2, random_state=42)
    emb_2d = tsne.fit_transform(best_embeddings)

    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(emb_2d[:, 0], emb_2d[:, 1], c=best_preds, cmap="tab10", alpha=0.7)
    plt.colorbar(scatter, ticks=range(10))
    plt.title("t-SNE of Best Matched Images (Zero-shot CLIP on MNIST)")
    plt.xlabel("t-SNE Dimension 1")
    plt.ylabel("t-SNE Dimension 2")
    plt.show()
else:
    print("Not enough high-confidence images for t-SNE visualization.")
```

# Explanation of Zero-shot CLIP Accuracy on MNIST using ViT-L/14

The zero-shot CLIP model achieved an accuracy of **55.48%** on the MNIST test set using the ViT-L/14 variant. While this is significantly above random chance (10% for 10 classes), it is lower than the performance of models specifically trained on MNIST. Several factors contribute to this result:

## Factors Affecting the Accuracy

1. **Domain Gap:**
   - **Training Data Mismatch:**  
     CLIP was pre-trained on a diverse dataset of natural images paired with text. In contrast, MNIST consists of simple, grayscale images of handwritten digits. This difference means that the features CLIP has learned to extract may not be ideal for distinguishing the subtle differences in handwritten digits.
   - **Feature Extraction Limitations:**  
     The ViT-L/14 encoder excels at capturing detailed information from complex images. However, when applied to MNIST, it may not extract the fine-grained features needed for optimal digit discrimination.

2. **Zero-shot Setup:**
   - **No Domain-specific Fine-tuning:**  
     In a zero-shot scenario, the model is used as-is without any fine-tuning on the MNIST dataset. As a result, the representations and text embeddings are not optimized for the characteristics of handwritten digits.
   - **Text Prompt Design:**  
     The text prompts used—such as “the handwritten digit zero,” “the handwritten digit one,” etc.—are fairly generic. Since these prompts are very similar to one another, their resulting text embeddings can be close in the embedding space, reducing the distinctiveness between classes.

3. **Similarity of Labels:**
   - **High Inter-class Similarity:**  
     The labels (and corresponding text embeddings) for digits are naturally similar because they describe numbers with subtle visual differences. This similarity in the textual domain can lead to overlapping or very close embeddings, which makes it more challenging for the model to accurately distinguish between classes.
   - **Cosine Similarity Challenges:**  
     When using cosine similarity to compare image embeddings to these similar text embeddings, small differences in the embeddings can lead to misclassification, further lowering overall accuracy.

## Visualization with t-SNE

To gain further insight, we visualized the embeddings of the "best matched" images using t-SNE:
- **Cluster Formation:**  
  The t-SNE plot can reveal how the embeddings cluster according to the model’s predictions. Well-defined clusters would suggest clear separation between classes.
- **Overlapping Clusters:**  
  Due to the similarity of the labels and the inherent challenge of the domain gap, clusters may overlap, indicating confusion between similar digits. This overlap helps explain the moderate accuracy of 55.48%.

## Summary

- **Accuracy Insight:**  
  The 55.48% accuracy reflects the challenges of applying a zero-shot approach to a domain (handwritten digits) that differs significantly from the natural images used to pre-train CLIP.
- **Impact of Similar Labels:**  
  The generic and highly similar nature of the text prompts results in text embeddings that are not sufficiently distinct, leading to classification confusion.
- **t-SNE Visualization:**  
  The t-SNE plots reveal that while there is some clustering of similar digits, significant overlap occurs—especially among classes with similar text descriptions—contributing to the observed accuracy.

In conclusion, although the zero-shot CLIP model shows promising generalization by achieving 55.48% accuracy without any fine-tuning, the accuracy is limited by the domain gap and the similarity of the class label embeddings.
